**`(?)`** 
1. Create a series from a numeric column that has the value of 'high' if it is equal to or above
the mean and 'low' if it is below the mean using .apply.
2. Create a series from a numeric column that has the value of 'high' if it is equal to or above
the mean and 'low' if it is below the mean using np.select.
3. Set the index to monotonically increasing integers starting from 0, then convert these to the
string version. Save this as s2.
4. Using s2, create a series with values with index entries '5', '3', and '2'.

```python
import pandas as pd
import numpy as np
```

```python
num_col = [10, 11, 15, 14, 8, 12, 10, 14]
srs = pd.Series(num_col)
```

```python
## 1
```

```python
def apply_func(val):
    if val >= srs.mean():
        return 'high'
    if val < srs.mean():
        return 'low' 
```

```python
srs.apply(apply_func)
```

```python
## 2
```

```python
"""
np.select(condlist, choicelist, default=0)

Docstring:
Return an array drawn from elements in choicelist, depending on conditions.

Parameters
----------
condlist : The list of conditions which determine from which array in `choicelist`
    the output elements are taken. When multiple conditions are satisfied,
    the first one encountered in `condlist` is used.
choicelist : The list of arrays from which the output elements are taken. It has
    to be of the same length as `condlist`.
default : scalar, optional. The element inserted in `output` when all conditions evaluate to False.
""";
```

```python
srs_ary = srs.to_numpy()
```

```python
condlist = [srs_ary >= srs_ary.mean(), srs_ary < srs_ary.mean()]
choicelist = ['high', 'low']
np.select(condlist, choicelist)
```

```python
## 3
```

```python
s2 = srs.reset_index().rename(index={x: str(x) for x in srs.index})
```

```python
s2.index
```

```python
## 4
```

```python
s2.loc[['5', '3', '2']]
```

**`(?)`**  Group by two categorical columns and take the mean of the numeric columns.

```python
# vehicle_df.info()
```

```python
# vehicle_df.drive.value_counts()
```

```python
# vehicle_df.cylinders.value_counts()
```

Upon inspection we can see that it is viable to convert the 'make' and 'drive' columns to categorical type. So, let's do that.

```python
vehicle_df = vehicle_df.astype({"make": "category", "drive": "category"})
```

```python
# using groupby()
vehicle_df.groupby(by=['make', 'drive']).agg(
    {col_numeric: "mean" for col_numeric in vehicle_df.select_dtypes("number").columns}).unstack()
```

```python
# using pivot_table()
vehicle_df.pivot_table(index='make', columns='drive', values=vehicle_df.select_dtypes("number"), aggfunc='mean').sort_index(axis='columns')
```

**`(?)`** With a dataset of your choice:
1. Add a new column that is the sum of a numeric column that was grouped by a string column.
2. Filter out the rows that have less than 3 entries when grouped by a string column.

```python
## 1

# num_car = number_of_cars_made_by_this_company
# vehicle_df.make.value_counts() as the new column, 'num_car'
vehicle_df_num_car = vehicle_df.assign(num_car = vehicle_df.groupby("make").make.transform('size'))
vehicle_df_num_car.set_index("make").num_car.sort_index()
```

```python
## 2

vehicle_df_num_car.groupby("make").filter(lambda grp_: grp_.make.agg('size') > 3)
```
